```python
#import all the needed libraries
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from scipy import random 

#limit of integration
a = 0
b = 1.75

#number of samples 
N = 20000

#define the function
def func(x): 
    return np.cos(x)

#define the values of cosine from 0 to 1.75
x_of_cos = np.linspace(0,1.75,10000)
y_of_cos = func(x_of_cos)

#define the min and max 
min_of_cos = min(y_of_cos)
max_of_cos = max(y_of_cos)

#creating random values
x_random = np.random.uniform(a,b,N)
y_random = np.random.uniform(min_of_cos,max_of_cos,N)

#classify where the point is
inside = np.where(np.cos(x_random)/y_random>1)[0]
outside = np.where(np.cos(x_random)/y_random<=1)[0]

#answer of the integral by just integrating cosine
integral_of_cos = float(np.sin(b) - np.sin(a))

#plotting the figure
figure = plt.figure()
plt.grid()
plt.xlim([a,b])
plt.ylim([min_of_cos,max_of_cos])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("F(0,1.75) = " + str(integral_of_cos))
plt.plot(x_random[inside],y_random[inside],".",color="blue",label="Points below f(x)")
plt.plot(x_random[outside],y_random[outside],".",color="grey",label="Points above f(x)")
plt.plot(x_of_cos,y_of_cos,color="red",label="f(x)")
plt.legend()

#estimating the value of the integral 
y_dimension = abs(min_of_cos - max_of_cos)
x_dimension = b - a 
area = y_dimension * x_dimension
value_of_integral = area*(len(inside)/float(N))

#calculating the error
error = (value_of_integral - np.sin(1.75))/np.sin(1.75)

#printing information
print("Number of samples: ",N)
print("Approximate value of the integral: ",value_of_integral)
print("Error in the approximation: ",error)
```
